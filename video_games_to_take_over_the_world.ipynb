```python
in_df = []
in_df2 = []
in_df3 = []
in_df4 = []
in_df5 = []
df_lowered = df.Title.str.lower()
df2_lowered = df2.Title.str.lower()
df3_lowered = df3.title.str.lower()
df4_lowered = df4.title.str.lower()
df5_lowered = df5.Name.str.lower().unique()
for game in df5_lowered:
    if game in df_lowered.values:
        in_df.append(1)
    else:
        in_df.append(0)
    
    if game in df2_lowered.values:
        in_df2.append(1)
    else:
        in_df2.append(0)
    
    if game in df3_lowered.values:
        in_df3.append(1)
    else:
        in_df3.append(0)
    
    if game in df4_lowered.values:
        in_df4.append(1)
    else:
        in_df4.append(0)
    
    in_df5.append(1)
```

```python
df5['single_player'] = df5.Stats.map(lambda x: x.get('Single-Player'))
df5.loc[df5.single_player.notna(), 'main_story'] = df5[df5.single_player.notna()].single_player.map(lambda x: x.get('Main Story'))
df5.loc[df5.main_story.notna(), 'main_average'] = df5[df5.main_story.notna()].main_story.map(lambda x: x.get('Average'))
df5.loc[df5.single_player.notna(), 'all_playstyles'] = df5[df5.single_player.notna()].single_player.map(lambda x: x.get('All PlayStyles'))
df5.loc[df5.all_playstyles.notna(), 'all_styles_average'] = df5[df5.all_playstyles.notna()].all_playstyles.map(lambda x: x.get('Average'))
df5.loc[df5.single_player.notna(), 'completionist'] = df5[df5.single_player.notna()].single_player.map(lambda x: x.get('Completionist'))
df5.loc[df5.completionist.notna(), 'completionist_average'] = df5[df5.completionist.notna()].completionist.map(lambda x: x.get('Average'))

df5['multi_player'] = df5.Stats.map(lambda x: x.get('Multi-Player'))
df5.loc[df5.multi_player.notna(), 'co_op'] = df5[df5.multi_player.notna()].multi_player.map(lambda x: x.get('Co-Op'))
df5.loc[df5.co_op.notna(), 'co_op_average'] = df5[df5.co_op.notna()].co_op.map(lambda x: x.get('Average'))
df5.loc[df5.multi_player.notna(), 'competitive'] = df5[df5.multi_player.notna()].multi_player.map(lambda x: x.get('Competitive'))
df5.loc[df5.competitive.notna(), 'competitive_average'] = df5[df5.competitive.notna()].competitive.map(lambda x: x.get('Average'))
```
